```python
class StockDataFetcher:
    def __init__(self, tickers, start_date, end_date, data_dir):
        self.tickers = tickers
        self.start_date = start_date
        self.end_date = end_date
        self.data_dir = data_dir

    def fetch_and_save(self):
        """Fetches stock data and saves as CSV."""
        for ticker in self.tickers:
            df = yf.download(ticker, start=self.start_date, end=self.end_date)
            print("Original DataFrame:")
            print(df.head())

            # Check if the DataFrame has a MultiIndex
            if isinstance(df.columns, pd.MultiIndex):
                df = df.stack(level=1, future_stack=True).reset_index()
                df.columns = ['Date', 'Ticker'] + list(df.columns[2:])
                print("Transformed DataFrame:")
                print(df.head())
            else:
                print("DataFrame does not have a MultiIndex. Skipping transformation.")

            file_path = self.data_dir + f"/{ticker}.csv"
            df.to_csv(file_path, index=False)
            print(f"Saved {ticker} data to {file_path}")

# Example usage
fetcher = StockDataFetcher(tickers=['AAPL', 'GOOG'], start_date='2020-01-01', end_date='2020-01-10', data_dir=r"C:\Users\Szabó Bendegúz\OneDrive - Ace Advisory\Documents\adatbank_rolinak\data")
fetcher.fetch_and_save()
```

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
from tqdm.notebook import tqdm
import pandas as pd
import time
from requests.adapters import HTTPAdapter, Retry
from itertools import repeat
from concurrent.futures import ThreadPoolExecutor
import concurrent.futures
import random
import re
from datetime import datetime, timedelta

def tm_profile_scraper_helper(player):
    ua = UserAgent()
    headers = {"User-Agent": ua.random}


    retries = 0
    while retries <= 3:
        try:
            tree = requests.get(player["link"], headers=headers)
            time.sleep(random.randint(1, 2))
            tree.raise_for_status()  # Raise an exception for 4xx or 5xx status codes
            
            break  # If successful, break out of retry loop
        except requests.HTTPError as e:
            print(f"HTTP error occurred while fetching URL: {player['link']}")
            print(f"Error message: {e}")
        except Exception as e:
            print(f"An error occurred while fetching URL: {player['link']}")
            print(f"Error message: {e}")
        retries += 1
    else:
        # If all retries fail, return an empty list
        print(f"Failed to fetch URL after {3} attempts: {player['link']}")
        return []

    soup = BeautifulSoup(tree.content, "html.parser")

    d = dict()
    d["league"] = player["league"]
    d["league_url"] = player["league_link"]
    d["tm_url"] = player["link"]
    d["tm_club"] = player["team"]
    d["tm_club_url"] = player["club_link"]
    d["tm_club_url"] = player["club_link"]
    # scrapes the name in the h1 tag
    d["tm_name"] = (
        soup.find("h1", {"class": "data-header__headline-wrapper"})
        .text.split("\n")[-1]
        .strip()
    )
    # d['tm_name'] = soup.find('h1', {'itemprop': 'name'}).text
    # scrapes the dataRibbon next to the club image for the transaction columns
    if soup.find("div", {"class": "data-header__ribbon"}) is not None:
        dataRibbon = soup.find("div", {"class": "data-header__ribbon"}).a[
            "title"
        ]
        if "Joined as a winter arrival from" in dataRibbon:
            d["tm_transaction_type"] = "Joined as a winter arrival from"
            d["tm_transaction_club"] = dataRibbon.split(";")[0].split(": ")[
                -1
            ]
            d["tm_transaction_fee"] = dataRibbon.split(";")[-1].split(": ")[
                -1
            ]
        elif "Returned after loan spell" in dataRibbon:
            d["tm_transaction_type"] = "Returned after loan spell"
            d["tm_transaction_club"] = dataRibbon.split(";")[0].split(
                " with "
            )[-1]
            d["tm_transaction_fee"] = dataRibbon.split(";")[-1].split(": ")[
                -1
            ]
        elif "On loan from" in dataRibbon:
            d["tm_transaction_type"] = "On loan from"
            d["tm_transaction_club"] = dataRibbon.split("On loan from ")[
                -1
            ].split(" until ")[0]
            d["tm_transaction_fee"] = "N/A"
        elif "Joined from" in dataRibbon:
            d["tm_transaction_type"] = "Joined from"
            d["tm_transaction_club"] = dataRibbon.split(";")[0].split(
                "Joined from "
            )[-1]
            d["tm_transaction_fee"] = dataRibbon.split(";")[-1].split(": ")[
                -1
            ]
        elif "Internal transfer" in dataRibbon:
            d["tm_transaction_type"] = "Internal transfer"
            d["tm_transaction_club"] = dataRibbon.split(";")[0].split(": ")[
                -1
            ]
            d["tm_transaction_fee"] = "N/A"
        elif "Joining" in dataRibbon:
            d["tm_transaction_type"] = "Joining"
            d["tm_transaction_club"] = dataRibbon.split(";")[0].split(": ")[
                -1
            ]
            d["tm_transaction_fee"] = dataRibbon.split(";")[-1].split(": ")[
                -1
            ]
        else:
            d["tm_transaction_type"] = dataRibbon
    else:
        d["tm_transaction_type"] = "N/A"
        d["tm_transaction_club"] = "N/A"
        d["tm_transaction_fee"] = "N/A"
    # scrapes the value in the blue box
    try:
        d["tm_value"] = soup.find(
            "div", {"class": "data-header__box--small"}
        ).a.text.split(" ")[0]
    except AttributeError:
        d["tm_value"] = "check"
    # scrapes the player data table for the relevant columns
    data_headers = soup.select(
        "div.info-table.info-table--right-space span.info-table__content.info-table__content--regular"
    )
    data_items = soup.select(
        "div.info-table.info-table--right-space span.info-table__content.info-table__content--bold"
    )
    if len(data_headers) != len(data_items):
        print(f"Profile table error: {player}")
    for header, item in zip(data_headers, data_items):
        header_clean = header.text
        if "Position" in header_clean:
            if "Goalkeeper" in item.text:
                d["tm_position"] = item.text.strip()
            else:
                d["tm_position"] = item.text.split(" - ")[-1].strip()
        elif "Joined" in header_clean:
            d["tm_transaction_date"] = item.text.strip()
        elif "Contract expires" in header_clean:
            d["tm_contract"] = item.text
        elif "Contract there expires" in header_clean:
            d["tm_contrac_until"] = item.text
        elif "Contract option" in header_clean:
            d["tm_contract_option"] = item.text
        elif "Citizenship" in header_clean:
            d["tm_nation1"] = item.text.split("\xa0\xa0")[1].strip()
            if len(item.text.split("\xa0\xa0")) > 2:
                d["tm_nation2"] = item.text.split("\xa0\xa0")[2].strip()
            else:
                d["tm_nation2"] = "N/A"
        # elif "Date of birth" in header_clean:
        #     d["tm_born"] = item.text
        elif "Date of birth/Age:" in header_clean:
            d["tm_born"] =  item.text.split('(')[0].strip() 
        elif "Full name" in header_clean:
            d["tm_fullname"] = item.text
        elif "Name in home country" in header_clean:
            d["tm_name_in_home_country"] = item.text
        elif "Foot" in header_clean:
            d["tm_foot"] = item.text
    # Scrapes the first row of the international table if it's present
    intl_table = soup.select("div.box .grid .national-career__row")

    if len(intl_table) > 1:  # Skip header row
        intl_row = intl_table[1]  # First data row

        # Get the national team name
        team_cell = intl_row.find("div", {"class": "grid__cell grid__cell--club"})
        if team_cell:
            d["tm_intl_team"] = team_cell.a["title"]
        
        # Get debut date (3rd column)
        debut_cell = intl_row.find_all("div", {"class": "grid__cell grid__cell--center"})[0]
        print(debut_cell.text)
        d["tm_intl_debut"] = debut_cell.text.strip() if debut_cell else "N/A"
        
        # Get the number of caps (4th column)
        caps_cell = intl_row.find_all("div", {"class": "grid__cell grid__cell--center"})[1]
        d["tm_intl_caps"] = caps_cell.text.strip() if caps_cell else "N/A"
        print(caps_cell.text)
        
        # Get the number of goals (5th column)
        goals_cell = intl_row.find_all("div", {"class": "grid__cell grid__cell--center"})[2]
        d["tm_intl_goals"] = goals_cell.text.strip() if goals_cell else "N/A"
        print(goals_cell.text)
    else:
        d["tm_intl_team"] = "N/A"
        d["tm_intl_debut"] = "N/A"
        d["tm_intl_caps"] = "N/A"
        d["tm_intl_goals"] = "N/A"

    # intl_table = soup.select('table.borderloser_odd_even_style tr')
    # if len(intl_table) > 0:
    #     d['tm_intl_team'] = intl_table[1].find('td', {'class': 'hauptlink no-border-links'}).a.text
    #     d['tm_intl_caps'] = intl_table[1].find_all('td', {'class': 'zentriert'})[3].text
    #     d['tm_intl_goals'] = intl_table[1].find_all('td', {'class': 'zentriert'})[4].text
    # else:
    #     d['tm_intl_team'] = 'N/A'
    #     d['tm_intl_caps'] = 'N/A'
    #     d['tm_intl_goals'] = 'N/A'
    # fills the missing colums with the appropriate values
    if "tm_fullname" not in d:
        d["tm_fullname"] = (
            soup.find("h1", {"class": "data-header__headline-wrapper"})
            .text.split("\n")[-1]
            .strip()
        )
    if "tm_contrac_until" not in d:
        d["tm_contrac_until"] = "N/A"
        
    return d
```
